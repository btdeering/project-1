```python
import pandas as pd
from pathlib import Path
```

```python

df = pd.read_csv("C:/Users/Marquis Sills/Desktop\DataBootCamp/project_1/education_per_state.csv")
df
# Remove rows containing the value 'x'
dfresult = df.dropna(how='all')

improved = dfresult[~(dfresult == '(X)').any(axis=1)]
improved
```

```python
# Create a boolean mask indicating which rows contain the phrase "!!Male!!Estimate"
male_mask = df['Label (Grouping)'].str.contains('!!Male!!Estimate')
female_mask = df['Label (Grouping)'].str.contains('!!Female!!Estimate')

# Filter the original dataframe into separate dataframes for male and female estimates
male_df = df[male_mask]
female_df = df[female_mask]

# Split the 'data' column into separate columns for each dataframe
male_df[['state', 'gender', 'estimate']] = male_df['Label (Grouping)'].str.split('!!', expand=True)
female_df[['state', 'gender', 'estimate']] = female_df['Label (Grouping)'].str.split('!!', expand=True)

# Drop the original 'data' column from each dataframe
male_df.drop('Label (Grouping)', axis=1, inplace=True)
female_df.drop('Label (Grouping)', axis=1, inplace=True)
male_df = male_df.replace('N', 0)
female_df = female_df.replace('N', 0)

```

```python
# Remove any leading or trailing whitespace characters from the column names
male_df.columns = male_df.columns.str.strip()
female_df.columns = female_df.columns.str.strip()
```

```python
#Output to EXCEL

male_df.to_csv("outputs/education_per_state_male_demographic.csv", index=False,encoding='utf-8-sig')
female_df.to_csv("outputs/education_per_state_female_demographic.csv", index=False, encoding='utf-8-sig')
```
