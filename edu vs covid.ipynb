```python
from pathlib import Path
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import statsmodels.formula.api as smf

```

```python
df = pd.read_csv("C:/Users/Marquis Sills/Desktop/DataBootCamp/project_1/education_covid.csv")
df.head()
```

```python
# Plot out education
x_values = df['high school/pop']*100
y_values = df['covid/pop']*100
plt.scatter(x_values,y_values)
plt.title(" High School Education and Covid Death")
plt.xlabel('Percent of High School Graduates Only')
plt.ylabel('Covid Death Percent')

# Print out the r-squared value along with the plot.
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,3)) + "x + " + str(round(intercept,3))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0.20,0.0006),fontsize=15,color="red")

rvalue_rounded = round(rvalue, 2)
plt.annotate(f"r: {rvalue_rounded}",(0.20,0.025),fontsize=15, color = "black")
print(f"The correlation is: {rvalue_rounded}")
plt.show()
```
