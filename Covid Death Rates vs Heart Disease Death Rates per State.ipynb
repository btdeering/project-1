```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.stats import linregress

data = pd.read_csv("C:\\Users\\btdee\\project-1\\data\\Heart_Covid.csv")

plt.scatter(data['Heart_Rate'], data['Covid_Rate'])
plt.xlabel('Heart Disease Mortality by State Per 100,000')
plt.ylabel('Covid-19 Death Rate Per 100,000')
plt.title('Covid Death Rate vs Heart Disease Mortality by State 2021')

slope, intercept, r_value, p_value, std_err = linregress(data['Heart_Rate'], data['Covid_Rate'])
plt.plot(data['Heart_Rate'], intercept + slope * data['Heart_Rate'], 'r')

equation = f'y = {slope:.2f}x + {intercept:.2f}\nR = {r_value:.2f}'
plt.text(0.5, 0.1, equation, transform=plt.gca().transAxes)

plt.show()
```
