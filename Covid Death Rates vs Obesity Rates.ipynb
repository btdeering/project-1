```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.read_csv("C:\\Users\\btdee\\project-1\\data\\Obesity_Covid.csv")

plt.scatter(data["Obesity_Rate"], data["Covid_Rate"], c='blue')

slope, intercept, r_value, p_value, std_err = stats.linregress(data["Obesity_Rate"], data["Covid_Rate"])
line = slope * data["Obesity_Rate"] + intercept
plt.plot(data["Obesity_Rate"], line, color="red")

plt.scatter(data["Obesity_Rate"], data["Covid_Rate"])

slope, intercept, r_value, p_value, std_err = stats.linregress(data["Obesity_Rate"], data["Covid_Rate"])
line = slope * data["Obesity_Rate"] + intercept
plt.plot(data["Obesity_Rate"], line, color="red")

plt.xlabel("Obesity Rate (%)")
plt.ylabel("Covid-19 Death Rate Per 100,000")
plt.title("Covid-19 Death Rates vs Obesity Rates per State 2021")

equation = f'y = {slope:.2f}x + {intercept:.2f}\nR = {r_value:.2f}'
plt.text(0.6, 0.1, equation, ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)

plt.show()
```
